```python
## append path
import sys; 
sys.path.append('src/')

## import packages 
import stats as st
import scipy.cluster.hierarchy as hr
import utils, viz
```

```python
## set results folder
fld = './yeast_dense_gamma/cont/'

## load pars.dat
pars = loadtxt(fld+'pars.txt')
eta  = loadtxt(fld+'eta.txt')

## sample every di steps and burnin
t0 = 0
te = 10000
di = 1

i0 = nonzero(pars[:,0] >= t0)[0][0]
ie = nonzero(pars[:,0] <= te)[0][-1]

## plot number of clusters
figure(figsize=(20,6)); 
subplot(121); plot(pars[i0:ie:di,0],pars[i0:ie:di,1]); xlabel('iteration #'); ylabel('estimated # of clusters'); grid(); #ylim([0,100]);  
subplot(122); hist(pars[i0:ie:di,1], 100, normed=True); xlabel('estimated # of clusters'); ylabel('density'); grid();

## plot remaining variables
figure(figsize=(20,6)); 
subplot(221); plot(eta[i0:ie:di,0],eta[i0:ie:di,1:]); xlabel('iteration #'); ylabel(r'$\eta_0$, $\eta_j$'); grid();
subplot(222); plot(pars[i0:ie:di,0],pars[i0:ie:di,2]); xlabel('iteration #'); ylabel(r'$\mu_\beta$'); grid(); #ylim([-15,0]); 
subplot(223); plot(pars[i0:ie:di,0],pars[i0:ie:di,3]); xlabel('iteration #'); ylabel(r'$\sigma^2_\beta$'); grid(); #ylim([0,50]);  
subplot(224); plot(pars[i0:ie:di,0],pars[i0:ie:di,[4,5]]); xlabel('iteration #'); ylabel(r'$\lambda_\alpha$'); grid(); #ylim([0,3]);  

## plot mean estimated hyperparameters
figure(figsize=(20,6)); 
subplot(121); x = linspace(-50,40,1000); plot(x, exp(st.dLogNormal(x, pars[i0:ie:di,2].mean(), pars[i0:ie:di,3].mean()))); xlabel(r'$\beta^*$'); ylabel('density'); grid(); xlim([-30, 10])
#subplot(122); x = linspace(0,5,1000); plot(x, exp(st.dLogGamma(x, shape = 0.75, scale=1/pars[i0:ie:di,4].mean()))); xlabel(r'$\alpha^*$'); ylabel('density'); grid();
subplot(122); x = linspace(0,5,1000); plot(x, exp(st.dLogGamma(x, shape = pars[i0:ie:di,4].mean(), scale=pars[i0:ie:di,5].mean()))); xlabel(r'$\alpha^*$'); ylabel('density'); grid();
```

```python
## compute gene posterior probabilities
pvalues = utils.computeGenePosteriorProbabilities(fld+'Z/', T0 = t0, T = te, step = di, idxs1 = [0,], idxs2 = [8,]) 

## visualize
figure(figsize=(10,6)); 
plot(pvalues,'.');
xlabel('genes'); ylabel('p');
```

```python
## check for DE genes
thr = 1.5
logodds = log10( (1 - pvalues) / pvalues );
ii  = logodds > thr
ids = arange(ii.size) 

print >> sys.stderr, 'Number of differentially expressed genes: {0}'.format(logodds[ii].size)

## compute FDR
jj  = pvalues < 1. / (10**thr + 1.)
fdr = pvalues[jj].sum() / jj.sum() 

print >> sys.stderr, 'False Discovery Rate: {0}'.format(fdr)

## visualize
figure(figsize=(10,6)); 
plot(ids[~ii], logodds[~ii],'.', ids[ii], logodds[ii],'r.');
xlabel('genes'); ylabel('log(posterior odds)');
```
