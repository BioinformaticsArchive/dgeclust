```python
## append path
import sys, os; 
sys.path.append('src/')

## import packages 
import pymc as mc
import scipy.cluster.hierarchy as hr
import stats, utils, viz
```

```python
## set results folder
fld = 'results/tmp3/'

## read pars.txt and eta.txt
eta  = loadtxt(os.path.join(fld,'eta.txt'))[:,1:]
pars = loadtxt(os.path.join(fld,'pars.txt'))

t  = pars[:,0]
N  = pars[:,1]
mu = pars[:,2]
s2 = pars[:,3]
sh = pars[:,4]
sc = pars[:,5]

## sampling interval
tini = 10
tend = 5000
di   = 10

ii = (t >= tini) & (t <= tend) & (t % di == 0) 

## plot chains
figure(figsize=(20,15)); 
subplot(4,2,1); hist(N[ii], 100, normed=True, color='grey'); xlabel('iteration #'); ylabel('clusters #');  
subplot(4,2,2); plot(t[ii], N[ii],            color='grey'); hlines(N[ii].mean(),  tini, tend, color='k', linestyle='--'); xlabel('iteration #'); ylabel('clusters #');  
subplot(4,2,3); plot(t[ii], mu[ii],           color='grey'); hlines(mu[ii].mean(), tini, tend, color='k', linestyle='--'); xlabel('iteration #'); ylabel('mean');  
subplot(4,2,4); plot(t[ii], s2[ii],           color='grey'); hlines(s2[ii].mean(), tini, tend, color='k', linestyle='--'); xlabel('iteration #'); ylabel('variance');  
subplot(4,2,5); plot(t[ii], sh[ii],           color='grey'); hlines(sh[ii].mean(), tini, tend, color='k', linestyle='--'); xlabel('iteration #'); ylabel('shape');  
subplot(4,2,6); plot(t[ii], sc[ii],           color='grey'); hlines(sc[ii].mean(), tini, tend, color='k', linestyle='--'); xlabel('iteration #'); ylabel('scale');  
subplot(4,2,7); plot(t[ii], eta[ii,1]);  xlabel('iteration #'); ylabel('eta0');  
subplot(4,2,8); plot(t[ii], eta[ii,2:]); xlabel('iteration #'); ylabel('eta');  

## plot mean estimated hyperparameters
figure(figsize=(20,6)); 
subplot(1,2,1); x = linspace(-50,40,1000); plot(x, exp(stats.dLogNormal(x, mu[ii].mean(), s2[ii].mean()))); xlabel(r'$\beta^*$'); ylabel('density');
subplot(1,2,2); x = linspace(0,2,1000);    plot(x, exp(stats.dLogGamma(x, shape=sh[ii].mean(), scale=sc[ii].mean()))); xlabel(r'$\phi^*$'); ylabel('density');
```

```python
## sampling interval
tini = 0
tend = 20000
di   = 1

## plot diagnostics
figure(figsize=(20,20)); 

## Geweke's test
subplot(5,2,1); scores = asarray(mc.geweke(N[ii]));  plot(scores[:,0],scores[:,1],'o'); hlines([-2,0,2], scores[0,0],scores[-1,0], color='k', linestyle='--'); ylim([-2.5, 2.5]); xlim([scores[0,0],scores[-1,0]]); xlabel('First iteration'); ylabel('Z-score for clusters #'); 
subplot(5,2,3); scores = asarray(mc.geweke(mu[ii])); plot(scores[:,0],scores[:,1],'o'); hlines([-2,0,2], scores[0,0],scores[-1,0], color='k', linestyle='--'); ylim([-2.5, 2.5]); xlim([scores[0,0],scores[-1,0]]); xlabel('First iteration'); ylabel('Z-score for mean'); 
subplot(5,2,5); scores = asarray(mc.geweke(s2[ii])); plot(scores[:,0],scores[:,1],'o'); hlines([-2,0,2], scores[0,0],scores[-1,0], color='k', linestyle='--'); ylim([-2.5, 2.5]); xlim([scores[0,0],scores[-1,0]]); xlabel('First iteration'); ylabel('Z-score for variance'); 
subplot(5,2,7); scores = asarray(mc.geweke(sh[ii])); plot(scores[:,0],scores[:,1],'o'); hlines([-2,0,2], scores[0,0],scores[-1,0], color='k', linestyle='--'); ylim([-2.5, 2.5]); xlim([scores[0,0],scores[-1,0]]); xlabel('First iteration'); ylabel('Z-score for shape'); 
subplot(5,2,9); scores = asarray(mc.geweke(sc[ii])); plot(scores[:,0],scores[:,1],'o'); hlines([-2,0,2], scores[0,0],scores[-1,0], color='k', linestyle='--'); ylim([-2.5, 2.5]); xlim([scores[0,0],scores[-1,0]]); xlabel('First iteration'); ylabel('Z-score for scale'); 

## auto-correlation
subplot(5,2,2);  acorr(N[ii],  normed=True, detrend=mlab.detrend_mean, usevlines=True, maxlags=100); xlabel('lag'); ylabel('autocorrelation for clusters #'); 
subplot(5,2,4);  acorr(mu[ii], normed=True, detrend=mlab.detrend_mean, usevlines=True, maxlags=100); xlabel('lag'); ylabel('autocorrelation for mean');
subplot(5,2,6);  acorr(s2[ii], normed=True, detrend=mlab.detrend_mean, usevlines=True, maxlags=100); xlabel('lag'); ylabel('autocorrelation for variance');
subplot(5,2,8);  acorr(sh[ii], normed=True, detrend=mlab.detrend_mean, usevlines=True, maxlags=100); xlabel('lag'); ylabel('autocorrelation for shape');
subplot(5,2,10); acorr(sc[ii], normed=True, detrend=mlab.detrend_mean, usevlines=True, maxlags=100); xlabel('lag'); ylabel('autocorrelation for scale');
```

```python
## compute gene posterior probabilities
pvalues = utils.computeGenePosteriorProbabilities(fld+'Z/', T0 = t0, T = te, step = di, idxs1 = [0,], idxs2 = [8,]) 

## visualize
figure(figsize=(10,6)); 
plot(pvalues,'.');
xlabel('genes'); ylabel('p');
```

```python
## check for DE genes
thr = 1.5
logodds = log10( (1 - pvalues) / pvalues );
ii  = logodds > thr
ids = arange(ii.size) 

print >> sys.stderr, 'Number of differentially expressed genes: {0}'.format(logodds[ii].size)

## compute FDR
jj  = pvalues < 1. / (10**thr + 1.)
fdr = pvalues[jj].sum() / jj.sum() 

print >> sys.stderr, 'False Discovery Rate: {0}'.format(fdr)

## visualize
figure(figsize=(10,6)); 
plot(ids[~ii], logodds[~ii],'.', ids[ii], logodds[ii],'r.');
xlabel('genes'); ylabel('log(posterior odds)');
```
