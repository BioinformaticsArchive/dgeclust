```python
%load_ext autoreload
%autoreload 2

from __future__ import division

import os
import numpy as np
import pylab as pl
import pandas as pd
import sklearn.metrics as metrics

%matplotlib inline
```

```python
## input
base = '/Users/dimitris/Repositories/simdata/'
sets = ['set1','set2','set3','set4','set5','set6'] 
reps = ['1rep','2rep','4rep']
samples = [1,2,3]

pl.figure(figsize=(3.34645669*2,8.85826772))
titles = ['B(0,1000)','B(500,500)','B(0,3000)','B(1500,1500)','P(500,500)', 'R(500,500)']
for i, s in enumerate(sets):
    ytrues = [np.loadtxt(os.path.join(base, s, 'ytrue'+str(sample)+'.txt')) for sample in samples]
    auc = []
    for r in reps:
        ## scores
        yscores_DGEclust = [1 - pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'.txt')).Posteriors.values.ravel() for sample in samples]
        yscores_DESeq = [1 - pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_DESeq.txt')).pval.values.ravel() for sample in samples]
        yscores_DESeq2 = [1 - pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_DESeq2.txt')).pvalue.values.ravel() for sample in samples]
        yscores_edgeR = [1 - pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_edgeR.txt')).PValue.values.ravel() for sample in samples]
        yscores_baySeq = [pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_baySeq.txt')).Likelihood.values.ravel() for sample in samples]

        for yscore in yscores_DESeq:
            yscore[np.isnan(yscore)]=1

        for yscore in yscores_DESeq2:
            yscore[np.isnan(yscore)]=1

        ## roc curves
        fpr_DGEclust, tpr_DGEclust, _ = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_DGEclust)])
        fpr_DESeq, tpr_DESeq, _ = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_DESeq)])
        fpr_DESeq2, tpr_DESeq2, _ = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_DESeq2)])
        fpr_edgeR, tpr_edgeR, _ = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_edgeR)])
        fpr_baySeq, tpr_baySeq, _ = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_baySeq)])
        
        auc_DGEclust = [metrics.auc(f, t) for f, t in zip(fpr_DGEclust, tpr_DGEclust)]
        auc_DESeq = [metrics.auc(f, t) for f, t in zip(fpr_DESeq, tpr_DESeq)]
        auc_DESeq2 = [metrics.auc(f, t) for f, t in zip(fpr_DESeq2, tpr_DESeq2)]
        auc_edgeR = [metrics.auc(f, t) for f, t in zip(fpr_edgeR, tpr_edgeR)]
        auc_baySeq = [metrics.auc(f, t) for f, t in zip(fpr_baySeq, tpr_baySeq)]
        
        ## append
        auc.append(auc_DGEclust)
        auc.append(auc_DESeq)
        auc.append(auc_DESeq2)
        auc.append(auc_edgeR)
        auc.append(auc_baySeq)
        auc.append([])
                
    ## plot boxplot
    pl.subplot(np.ceil(len(sets)/2), 2, i+1); 
    bxp = pl.boxplot(auc, patch_artist=True); 
    colors = ['pink', 'lightblue', 'lightgreen', 'tan', 'cyan'] * 3 
    for box, color in zip(bxp['boxes'], colors):
        # change outline color
#         box.set( color='#7570b3', linewidth=2)
        # change fill color
        box.set( facecolor = color )

    ## change color and linewidth of the whiskers
    for whisker in bxp['whiskers']:
        whisker.set(color='k')

    ## change color and linewidth of the caps
    for cap in bxp['caps']:
        cap.set(color='k')

#     ## change color and linewidth of the medians
#     for median in bp['medians']:
#         median.set(color='#b2df8a', linewidth=2)

#     ## change the style of fliers and their fill
#     for flier in bp['fliers']:
#         flier.set(marker='o', color='#e7298a', alpha=0.5)  
    pl.vlines([6, 12], 0, 1, colors='k', linestyles='dashed')
    pl.xticks(range(1, 18), 
              ['A1', 'B1', 'C1', 'D1', 'E1',
               '',
               'A2', 'B2', 'C2', 'D2', 'E2',
               '',
               'A4', 'B4', 'C4', 'D4', 'E4']); 
    pl.ylim([0.5, 1])
    pl.ylabel('AUC', fontsize=10)
    pl.title(titles[i], fontsize=12)
    pl.setp(pl.gca().get_xticklabels(), fontsize=8)
    pl.setp(pl.gca().get_yticklabels(), fontsize=8)
    
## legend
pl.subplot(np.ceil(len(sets)/2), 2, 5)
pl.text(0.5, 0.45, 
        'A#: DGEclust\nB#: DESeq\nC#: DESeq2\nD#: edgeR\nE#: baySeq\n#: no of replicas', 
        transform=pl.gca().transAxes, 
        fontsize=10, 
        verticalalignment='top', 
        bbox=dict(facecolor='white', alpha=0.8))

##
pl.tight_layout()
pl.savefig('auc.pdf')
```

```python
## input
base = '/Users/dimitris/Repositories/simdata/'
sets = ['set1','set2','set3','set4','set5','set6'] 
reps = ['1rep','2rep','4rep']
samples = [1,2,3]
ngenes = 1000

pl.figure(figsize=(3.34645669*2,8.85826772))
titles = ['B(0,1000)','B(500,500)','B(0,3000)','B(1500,1500)','P(500,500)','R(500,500)']
for i, s in enumerate(sets):
    ytrues = [np.loadtxt(os.path.join(base, s, 'ytrue'+str(sample)+'.txt')) for sample in samples]
    fp_mean = []
    for r in reps:
        ## scores
        yscores_DGEclust = [pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'.txt')).Posteriors.values.ravel() for sample in samples]
        yscores_DESeq = [pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_DESeq.txt')).pval.values.ravel() for sample in samples]
        yscores_DESeq2 = [pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_DESeq2.txt')).pvalue.values.ravel() for sample in samples]
        yscores_edgeR = [pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_edgeR.txt')).PValue.values.ravel() for sample in samples]
        yscores_baySeq = [1-pd.read_table(os.path.join(base, s, r, 'pvals'+str(sample)+'_baySeq.txt')).Likelihood.values.ravel() for sample in samples]

        for yscore in yscores_DESeq:
            yscore[np.isnan(yscore)]=1

        for yscore in yscores_DESeq2:
            yscore[np.isnan(yscore)]=1
        
        ## sort
        idxs_DGEclust = [np.argsort(yscore) for yscore in yscores_DGEclust] 
        idxs_DESeq = [np.argsort(yscore) for yscore in yscores_DESeq] 
        idxs_DESeq2 = [np.argsort(yscore) for yscore in yscores_DESeq2] 
        idxs_edgeR = [np.argsort(yscore) for yscore in yscores_edgeR] 
        idxs_baySeq = [np.argsort(yscore) for yscore in yscores_baySeq] 
        
        ## false positives
        fp_DGEclust = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_DGEclust, ytrues)]
        fp_DESeq    = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_DESeq, ytrues)]
        fp_DESeq2   = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_DESeq2, ytrues)]
        fp_edgeR    = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_edgeR, ytrues)]
        fp_baySeq   = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_baySeq, ytrues)]
        
        ## averages
        fp_mean.append([np.mean(fp_DGEclust, 0), 
                        np.mean(fp_DESeq, 0), 
                        np.mean(fp_DESeq2, 0), 
                        np.mean(fp_edgeR, 0), 
                        np.mean(fp_baySeq, 0)])
        
    ## plot boxplot
    pl.subplot(np.ceil(len(sets)/2), 2, i+1);  
    for j, el in enumerate(fp_mean):
        x = range(j*1000, j*1000+ngenes)
        pl.plot(x, el[0], color='red') 
        pl.plot(x, el[1], color='blue')
        pl.plot(x, el[2], color='green')
        pl.plot(x, el[3], color='brown')
        pl.plot(x, el[4], color='cyan')     
    pl.vlines([1000, 2000], 1, 1000, colors='k', linestyles='dashed')
    pl.yscale('log')
    pl.ylim([1, 1000])
    pl.xticks([0,500,1000,1500, 2000, 2500], [0, 500, 0, 500, 0, 500])
    pl.text(0.02, 0.9, 'n=1', transform=pl.gca().transAxes, fontsize=10)
    pl.text(0.35, 0.9, 'n=2', transform=pl.gca().transAxes, fontsize=10)
    pl.text(0.68, 0.9, 'n=4', transform=pl.gca().transAxes, fontsize=10)
    pl.xlabel('number of positives')
    pl.ylabel('number of false positives', fontsize=10)    
    pl.title(titles[i], fontsize=12)
    pl.setp(pl.gca().get_xticklabels(), fontsize=8)
    pl.setp(pl.gca().get_yticklabels(), fontsize=8)

## legend
pl.subplot(np.ceil(len(sets)/2), 2, 1)
pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=(0.05,0.05), fontsize=10, framealpha=0.8)

##
pl.tight_layout()
pl.savefig('fdc.pdf')
```
