```python
## compute gene posterior probabilities
idxs1 = [1,2,4,5]
idxs2 = [0,3]
post = ut.computeGenePosteriorProbabilities(res, T0 = 40000, T = 300000, dt = 10, idxs1 = idxs1, idxs2 = idxs2) 
```

```python
## check for DE genes
thr = 1.
logodds = log10( (1 - post) / post );
ii  = logodds > thr
ids = arange(ii.size) 

print >> sys.stderr, 'Number of differentially expressed genes: {0}'.format(logodds[ii].size)

## compute FDR
jj  = post < 1. / (10**thr + 1.)
fdr = post[jj].sum() / jj.sum() 

print >> sys.stderr, 'False Discovery Rate: {0}'.format(fdr)

## visualize
figure(figsize=(20,6)); 
subplot(1,2,1); plot(ids[~ii], logodds[~ii],'k.', ids[ii], logodds[ii],'r.'); xlabel('genes'); ylabel('log(posterior odds)');
subplot(1,2,2); A, R = vz.plotRA(data.counts.T[idxs1], data.counts.T[idxs2], ii);
```
