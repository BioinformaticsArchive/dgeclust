```python
def plot_auc(base, sets, reps, samples):  
    pl.figure(figsize=(fig_width_max,fig_height_max))
    titles = ['10% DE genes','30% DE genes','10% DE genes + partial over-dispersion', '10% DE genes + outliers']
    for i, s in enumerate(sets):
        ytrues = [np.loadtxt(os.path.join(base, s, 'data', 'ytrue'+str(sample)+'.txt')) for sample in samples]
        auc = []
        for r in reps:
            ## scores
            yscores_DGEclust = [1 - pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'.txt')).Posteriors.values.ravel() for sample in samples]
            yscores_DESeq = [1 - pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_DESeq.txt')).pval.values.ravel() for sample in samples]
            yscores_DESeq2 = [1 - pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_DESeq2.txt')).pvalue.values.ravel() for sample in samples]
            yscores_edgeR = [1 - pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_edgeR.txt')).PValue.values.ravel() for sample in samples]
            yscores_baySeq = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_baySeq.txt')).Likelihood.values.ravel() for sample in samples]

            for yscore in yscores_DESeq:
                yscore[np.isnan(yscore)]=1

            for yscore in yscores_DESeq2:
                yscore[np.isnan(yscore)]=1

            ## roc curves
            fpr_DGEclust, tpr_DGEclust, thr_DGEclust = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_DGEclust)])
            fpr_DESeq, tpr_DESeq, thr_DESeq = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_DESeq)])
            fpr_DESeq2, tpr_DESeq2, thr_DESeq2 = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_DESeq2)])
            fpr_edgeR, tpr_edgeR, thr_edgeR = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_edgeR)])
            fpr_baySeq, tpr_baySeq, thr_baySeq = zip(*[metrics.roc_curve(ytrue, yscore) for ytrue, yscore in zip(ytrues, yscores_baySeq)])

            auc_DGEclust = [metrics.auc(f[thr<1], t[thr<1]) for f, t, thr in zip(fpr_DGEclust, tpr_DGEclust, thr_DGEclust)]
            auc_DESeq = [metrics.auc(f[thr<1], t[thr<1]) for f, t, thr in zip(fpr_DESeq, tpr_DESeq, thr_DESeq)]
            auc_DESeq2 = [metrics.auc(f[thr<1], t[thr<1]) for f, t, thr in zip(fpr_DESeq2, tpr_DESeq2, thr_DESeq2)]
            auc_edgeR = [metrics.auc(f[thr<1], t[thr<1]) for f, t, thr in zip(fpr_edgeR, tpr_edgeR, thr_edgeR)]
            auc_baySeq = [metrics.auc(f[thr<1], t[thr<1]) for f, t, thr in zip(fpr_baySeq, tpr_baySeq, thr_baySeq)]

            ## append
            auc.append(auc_DGEclust)
            auc.append(auc_DESeq)
            auc.append(auc_DESeq2)
            auc.append(auc_edgeR)
            auc.append(auc_baySeq)
            auc.append([])

        ## plot boxplot
        ax = pl.subplot(len(sets), 1, i+1); 
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()  
        ax.tick_params(axis='both', direction='out')

        bxp = pl.boxplot(auc, patch_artist=True); 
        colors = ['red', 'blue', 'green', 'brown', 'cyan', 'black'] * 3
        for box, color in zip(bxp['boxes'], colors):
            # change outline color
    #         box.set( color='#7570b3', linewidth=2)
            # change fill color
            box.set( facecolor = color )

        ## change color and linewidth of the whiskers
        for whisker in bxp['whiskers']:
            whisker.set(color='k')

        ## change color and linewidth of the caps
        for cap in bxp['caps']:
            cap.set(color='k')

        ## change color and linewidth of the medians
        for median in bxp['medians']:
            median.set(color='black', linewidth=0.5)

    #     ## change the style of fliers and their fill
    #     for flier in bp['fliers']:
    #         flier.set(marker='o', color='#e7298a', alpha=0.5)  
        pl.axvline(6, color='k', linestyle='dashed')
        pl.axvline(12, color='k', linestyle='dashed')
        pl.xticks(range(1, 18), 
                  ['A2', 'B2', 'C2', 'D2', 'E2',
                   '',
                   'A4', 'B4', 'C4', 'D4', 'E4',
                   '',
                   'A8', 'B8', 'C8', 'D8', 'E8']); 
        pl.ylim([0.6, 1.01])
        pl.ylabel('Area Under ROC Curve')
        pl.title(titles[i])

    ## legend
    pl.subplot(len(sets), 1, 1)
    pl.text(0.80, 0.6, 
            'A#: DGEclust\nB#: DESeq\nC#: DESeq2\nD#: edgeR\nE#: baySeq\n#: no of replicas', 
            transform=pl.gca().transAxes, 
            verticalalignment='top', 
            bbox=dict(facecolor='white', alpha=0.8))

##
base = '/Users/dimitris/Repositories/benchmarks/'
sets = ['set1','set3','set7','set8'] 
reps = ['2','4','8']
samples = [1,2,3]
plot_auc(base, sets, reps, samples)
pl.tight_layout()
# pl.savefig('auc.pdf')
```

```python
def plot_fdc(base, sets, reps, samples, ngenes=1000):
    pl.figure(figsize=(fig_width_max,fig_height_max))
    titles = ['10% DE genes','30% DE genes','10% DE genes + partial over-dispersion', '10% DE genes + outliers']
    for i, s in enumerate(sets):
        ytrues = [np.loadtxt(os.path.join(base, s, 'data', 'ytrue'+str(sample)+'.txt')) for sample in samples]
        fp_mean = []
        for r in reps:
            ## scores
            yscores_DGEclust = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'.txt')).Posteriors.values.ravel() for sample in samples]
            yscores_DESeq = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_DESeq.txt')).pval.values.ravel() for sample in samples]
            yscores_DESeq2 = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_DESeq2.txt')).pvalue.values.ravel() for sample in samples]
            yscores_edgeR = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_edgeR.txt')).PValue.values.ravel() for sample in samples]
            yscores_baySeq = [1-pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_baySeq.txt')).Likelihood.values.ravel() for sample in samples]

            for yscore in yscores_DESeq:
                yscore[np.isnan(yscore)]=1

            for yscore in yscores_DESeq2:
                yscore[np.isnan(yscore)]=1

            ## sort
            idxs_DGEclust = [np.argsort(yscore) for yscore in yscores_DGEclust] 
            idxs_DESeq = [np.argsort(yscore) for yscore in yscores_DESeq] 
            idxs_DESeq2 = [np.argsort(yscore) for yscore in yscores_DESeq2] 
            idxs_edgeR = [np.argsort(yscore) for yscore in yscores_edgeR] 
            idxs_baySeq = [np.argsort(yscore) for yscore in yscores_baySeq] 

            ## false positives
            fp_DGEclust = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_DGEclust, ytrues)]
            fp_DESeq    = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_DESeq, ytrues)]
            fp_DESeq2   = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_DESeq2, ytrues)]
            fp_edgeR    = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_edgeR, ytrues)]
            fp_baySeq   = [np.cumsum(ytrue != ytrue[idxs])[:ngenes] for idxs, ytrue in zip(idxs_baySeq, ytrues)]

            ## averages
            fp_mean.append([np.mean(fp_DGEclust, 0), 
                            np.mean(fp_DESeq, 0), 
                            np.mean(fp_DESeq2, 0), 
                            np.mean(fp_edgeR, 0), 
                            np.mean(fp_baySeq, 0)])

        ## plot boxplot
        ax = pl.subplot(len(sets), 1, i+1);  
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()  
        ax.tick_params(axis='both', direction='out')
        
        for j, el in enumerate(fp_mean):
            x = range(j*1000, j*1000+ngenes)
            pl.plot(x, el[0], color='red') 
            pl.plot(x, el[1], color='blue')
            pl.plot(x, el[2], color='green')
            pl.plot(x, el[3], color='brown')
            pl.plot(x, el[4], color='cyan')     
        pl.axvline(1000, color='k', linestyle='dashed')
        pl.axvline(2000, color='k', linestyle='dashed')

        pl.yscale('log')
        pl.ylim([1, 1000])
        pl.xticks([0,500,1000,1500,2000,2500], [0, 500, 0, 500, 0, 500])
        pl.text(0.01, 0.9, 'n=2', transform=pl.gca().transAxes)
        pl.text(0.34, 0.9, 'n=4', transform=pl.gca().transAxes)
        pl.text(0.67, 0.9, 'n=8', transform=pl.gca().transAxes)
        pl.xlabel('# of discoveries')
        pl.ylabel('# of false discoveries')    
        pl.title(titles[i])

    ## legend
    pl.subplot(len(sets), 1, 1)
    pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=(0.14,0.0), framealpha=0.8)

##
base = '/Users/dimitris/Repositories/benchmarks/'
sets = ['set1','set3','set7','set8'] 
reps = ['2','4','8']
samples = [1,2,3,2]
plot_fdc(base, sets, reps, samples)
pl.tight_layout()
# pl.savefig('fdc.pdf')
```

```python
## input
def plot_fdr(base, sets, reps, samples, thr):
    pl.figure(figsize=(fig_width_max,fig_height_max))
    titles = ['10% DE genes','30% DE genes','10% DE genes + partial over-dispersion', '10% DE genes + outliers']
    for i, s in enumerate(sets):
        ytrues = [np.loadtxt(os.path.join(base, s, 'data', 'ytrue'+str(sample)+'.txt')) for sample in samples]
        fdr = []
        for r in reps:
            ## scores
            yscores_DGEclust = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'.txt')).FDR.values.ravel() for sample in samples]
            yscores_DESeq = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_DESeq.txt')).padj.values.ravel() for sample in samples]
            yscores_DESeq2 = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_DESeq2.txt')).padj.values.ravel() for sample in samples]
            yscores_edgeR = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_edgeR.txt')).FDR.values.ravel() for sample in samples]
            yscores_baySeq = [pd.read_table(os.path.join(base, s, r, '2', 'pvals'+str(sample)+'_baySeq.txt'))['FDR.DE'].values.ravel() for sample in samples]

            ## idxs
            idxs_DGEclust = [yscore < thr for yscore in yscores_DGEclust]
            idxs_DESeq = [yscore < thr for yscore in yscores_DESeq]
            idxs_DESeq2 = [yscore < thr for yscore in yscores_DESeq2]
            idxs_edgeR = [yscore < thr for yscore in yscores_edgeR]
            idxs_baySeq = [yscore < thr for yscore in yscores_baySeq]        

            ##
            fdr_DGEclust = [np.sum(idxs[np.logical_not(ytrue)]) / np.sum(idxs) for idxs, ytrue in zip(idxs_DGEclust, ytrues)]
            fdr_DESeq = [np.sum(idxs[np.logical_not(ytrue)]) / np.sum(idxs) for idxs, ytrue in zip(idxs_DESeq, ytrues)]
            fdr_DESeq2 = [np.sum(idxs[np.logical_not(ytrue)]) / np.sum(idxs) for idxs, ytrue in zip(idxs_DESeq2, ytrues)]
            fdr_edgeR = [np.sum(idxs[np.logical_not(ytrue)]) / np.sum(idxs) for idxs, ytrue in zip(idxs_edgeR, ytrues)]
            fdr_baySeq = [np.sum(idxs[np.logical_not(ytrue)]) / np.sum(idxs) for idxs, ytrue in zip(idxs_baySeq, ytrues)]        

            ## averages
            fdr.append(fdr_DGEclust)
            fdr.append(fdr_DESeq)
            fdr.append(fdr_DESeq2)
            fdr.append(fdr_edgeR)
            fdr.append(fdr_baySeq)
            fdr.append([])

        ## plot boxplot
        ax = pl.subplot(len(sets), 1, i+1);  
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()  
        ax.tick_params(axis='both', direction='out')

        bxp = pl.boxplot(fdr, patch_artist=True); 
        colors = ['red', 'blue', 'green', 'brown', 'cyan','black'] * 3
        for box, color in zip(bxp['boxes'], colors):
            # change outline color
    #         box.set( color='#7570b3', linewidth=2)
            # change fill color
            box.set( facecolor = color )

        ## change color and linewidth of the whiskers
        for whisker in bxp['whiskers']:
            whisker.set(color='k')

        ## change color and linewidth of the caps
        for cap in bxp['caps']:
            cap.set(color='k')

        ## change color and linewidth of the medians
        for median in bxp['medians']:
            median.set(color='black', linewidth=0.5)

        ## change the style of fliers and their fill
        for flier in bxp['fliers']:
            flier.set(marker='+', color='k')    

        pl.axhline(thr, color='k', linestyle='dashed')
        pl.xticks(range(1, 18), 
                  ['A2', 'B2', 'C2', 'D2', 'E2',
                   '',
                   'A4', 'B4', 'C4', 'D4', 'E4',
                   '',
                   'A8', 'B8', 'C8', 'D8', 'E8']); 
        pl.ylim([-0.1, 1])
        pl.ylabel('FDR')
        pl.title(titles[i])

    ## legend
    pl.subplot(len(sets), 1, 1);      
    pl.text(0.80, 0.96, 
            'A#: DGEclust\nB#: DESeq\nC#: DESeq2\nD#: edgeR\nE#: baySeq\n#: no of replicas', 
            transform=pl.gca().transAxes, 
            verticalalignment='top', 
            bbox=dict(facecolor='white', alpha=0.8))

##
base = '/Users/dimitris/Repositories/benchmarks/'
sets = ['set1','set3','set7','set8'] 
reps = ['2','4','8']
samples = [1,2,3]
thr = 0.1
plot_fdr(base, sets, reps, samples, thr)
pl.tight_layout()
# pl.savefig('fdr.pdf')
```

```python
## Copyright (C) 2012-2013 Dimitrios V. Vavoulis
## Computational Genomics Group (http://bioinformatics.bris.ac.uk/)
## Department of Computer Science
## University of Bristol

## NOTE: Based on code found in the following link: http://stackoverflow.com/questions/2455761/reordering-matrix-elements-to-reflect-column-and-row-clustering-in-naiive-python/3017704#3017704

################################################################################

def plotHeatmap(Dvec, method = 'average', metric = 'euclidean', xticklabels = None, yticklabels = None, cmap = pl.cm.coolwarm): 
    grid = gr.GridSpec(2, 2, width_ratios=[1,3.5], height_ratios=[1,3.5], wspace=0, hspace=0)

    D = hr.distance.squareform(Dvec)
    D = (D - D.min()) / (D.max() - D.min())

    # Compute and plot y-axis dendrogram
    ax1 = pl.subplot(grid[2]); 
    Y1  = hr.linkage(Dvec, method = method, metric = metric);
    Z1  = hr.dendrogram(Y1, orientation = 'right')
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.set_frame_on(False)
                
    # Compute and plot x-axis dendrogram
    ax2 = pl.subplot(grid[1])
    Y2  = Y1 
    Z2  = hr.dendrogram(Y2)
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.set_frame_on(False)

    # Plot distance matrix
    axmatrix = pl.subplot(grid[3])
    idx1 = Z1['leaves']
    idx2 = Z2['leaves']
    yticklabels = np.asarray(yticklabels)[idx2].tolist()
    xticklabels = np.asarray(xticklabels)[idx1].tolist()
    D = D[idx1,:]
    D = D[:,idx2]
    
    im = axmatrix.matshow(D, aspect = 'auto', origin = 'lower', cmap = cmap)
    axmatrix.set_xticks(range(len(xticklabels)))
    axmatrix.set_yticks(range(len(yticklabels)))
    axmatrix.xaxis.set_tick_params(which = 'major', size=0)
    axmatrix.yaxis.set_tick_params(which = 'major', size=0)
    axmatrix.xaxis.set_ticks_position('bottom')
    axmatrix.yaxis.set_ticks_position('right')        
    axmatrix.set_xticklabels(xticklabels, rotation = 'vertical') 
    axmatrix.set_yticklabels(yticklabels)
    axmatrix.set_frame_on = False
    
    # Plot colorbar
    axcolor = pl.gcf().add_axes([0.07,0.9,0.12,0.05])
    pl.colorbar(im, cax = axcolor, orientation = 'horizontal', ticks = (0,1))
    
    
####
simvec = np.loadtxt('/Users/dimitris/Repositories/benchmarks/cage/2/DGEclust/simvec_groups.txt')

pl.figure(figsize=(fig_width_max/2,fig_width_max/2))
plotHeatmap(1-simvec, method = 'complete', metric = 'euclidean',
            xticklabels = ['Dorsal Striatum','Cerebral Cortex','Hippocampus','Dorsal Striatum','Cerebral Cortex'], 
            yticklabels = ['Caudate Nucleus','Frontal Gyrus','Hippocampus','Putamen','Temporal Gyrus'])

pl.tight_layout()
# pl.savefig('heatmap.pdf')
```

```python
def plot_rocs():    
    base = '/Users/dimitris/Repositories/benchmarks'
    orgs = ['mouse', 'rat', 'worm', 'fly', 'human_cage_2', 'human_cage_5']
    titles = ['Mouse (n=2)', 'Rat (n=2)', 'Worm (n=2)', 'Fly (n=3)', 
              'Human (caudate/frontal, n=2)', 'Human (caudate/frontal, n=5)']
    
    pl.figure(figsize=(fig_width_max, golden_ratio * fig_width_max))
    for i, org in enumerate(orgs):
        ax = pl.subplot(3,2,i+1)
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()  
        ax.tick_params(axis='both', direction='out')

        ytrues = np.loadtxt(os.path.join(base, org, 'ytrues.txt'))
        idxs = ytrues != 2
        ytrues = ytrues[idxs]
        
        pvals_DGEclust = pd.read_table(os.path.join(base, org, 'pvals_DGEclust.txt'), index_col=0)[idxs]
        pvals_DESeq = pd.read_table(os.path.join(base, org, 'pvals_DESeq.txt'), index_col=0)[idxs]
        pvals_DESeq2 = pd.read_table(os.path.join(base, org, 'pvals_DESeq2.txt'), index_col=0)[idxs]
        pvals_edgeR = pd.read_table(os.path.join(base, org, 'pvals_edgeR.txt'), index_col=0)[idxs]
        pvals_baySeq = pd.read_table(os.path.join(base, org, 'pvals_baySeq.txt'), index_col=0)[idxs]

        ## DGEclust
        fpr, tpr, _ = metrics.roc_curve(ytrues, 1-pvals_DGEclust.Posteriors)
        pl.plot(fpr, tpr, color='red');

        ## DESeq
        pvals_DESeq.pval[np.isnan(pvals_DESeq.pval)] = 1
        fpr, tpr, _ = metrics.roc_curve(ytrues, 1-pvals_DESeq.pval)
        pl.plot(fpr, tpr, color='blue');

        ## DESeq2 
        pvals_DESeq2.pvalue[np.isnan(pvals_DESeq2.pvalue)] = 1
        pvals_DESeq2.padj[np.isnan(pvals_DESeq2.padj)] = 1
        fpr, tpr, _ = metrics.roc_curve(ytrues, 1-pvals_DESeq2.pvalue)
        pl.plot(fpr, tpr, color='green');

        ## edgeR
        fpr, tpr, _ = metrics.roc_curve(ytrues, 1-pvals_edgeR.PValue)
        pl.plot(fpr, tpr, color='brown');

        ## baySeq
        fpr, tpr, _ = metrics.roc_curve(ytrues, pvals_baySeq.Likelihood)
        pl.plot(fpr, tpr, color='cyan');
        
        pl.title(titles[i])
        pl.xlabel('False Positive Rate')
        pl.ylabel('True Positive Rate')            

    pl.subplot(3,2,1)
    pl.xlim([0, 0.1]); pl.ylim([0.7, 1.01])
    pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=0, framealpha=1)

    pl.subplot(3,2,2)
    pl.xlim([0, 0.3]); pl.ylim([0.7, 1.01])
    pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=0, framealpha=1)

    pl.subplot(3,2,3)
    pl.xlim([0, 0.1]); pl.ylim([0.7, 1.01])
    pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=0, framealpha=1)

    pl.subplot(3,2,4)
    pl.xlim([0, 0.1]); pl.ylim([0.9, 1.01])
    pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=0, framealpha=1)

    pl.subplot(3,2,5)
    pl.xlim([0, 0.3]); pl.ylim([0.7, 1.01])
    pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=0, framealpha=1)

    pl.subplot(3,2,6)
    pl.xlim([0, 0.3]); pl.ylim([0.7, 1.01])
    pl.legend(['DGEclust', 'DESeq', 'DESeq2', 'edgeR', 'baySeq'], loc=4, framealpha=1)
    
##
plot_rocs()
pl.tight_layout()
pl.savefig('compare.pdf')
# pl.savefig('compare.png', transparent=True)
```
